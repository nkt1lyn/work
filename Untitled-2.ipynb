```python
import logging
import datetime
import telebot
import sqlite3
from telebot import types
from telegram.ext import Updater, CommandHandler, MessageHandler, ConversationHandler, CallbackContext


# Настройте логирование
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logger = logging.getLogger(__name__)

# Состояния для регистрации студентов
STUDENT_ID = range(1)

# Словарь для хранения данных о студентах (вы можете использовать базу данных)
students = {}

# Структура расписания занятий (день -> список занятий)
schedule = {
    'Понедельник': {
        'upper': [
            'Математика, 10:00, ауд. 101',
            'Физика, 14:00, ауд. 102'
        ],
        'lower': [
            'Информатика, 09:00, ауд. 103'
        ]
    },
    'Вторник': {
        'upper': [
            # Add lessons for Tuesday of the upper week
        ],
        'lower': [
            # Add lessons for Tuesday of the lower week
        ]
    },
    # Add other days and lessons
}

# Создание бота
bot = telebot.TeleBot('6408829408:AAFn9dtKi0WjImLxVL4BWYAuzNFUWKd_7ag')

# Обработчик команды /start
@bot.message_handler(commands=['start'])
def start(message):
    markup = types.ReplyKeyboardRemove(selective=False)
    bot.send_message(message.chat.id, "Привет! Я бот для регистрации студентов. "
                                      "Пожалуйста, отправь свой номер студенческого билета:",
                     reply_markup=markup)
    return STUDENT_ID

# Обработчик ввода номера студенческого билета
@bot.message_handler(func=lambda message: message.text.isdigit(), content_types=['text'])
def get_student_id(message):
    user_id = message.from_user.id
    student_id = message.text
    # Check if the student is already registered
    if student_id in students:
        bot.send_message(message.chat.id, "Вы уже зарегистрированы.")
        return
    students[user_id] = {'student_id': student_id}
    save_student_data(student_id, message.text)  # Save student data to SQLite
    bot.send_message(message.chat.id, f"Спасибо, ваш номер студенческого билета {student_id} сохранен.")
    show_schedule(message.chat.id)  # Показать расписание после регистрации



# Обработчик команды /schedule для просмотра расписания
@bot.message_handler(commands=['schedule'])
def show_schedule(chat_id):
    if chat_id in students:
        student_id = students[chat_id]['student_id']
        if student_id:  # Проверка, что студент зарегистрирован
            day = get_current_day()  # Получить текущий день недели
            week_type = get_current_week_type()  # Получить тип недели (верхняя или нижняя)
            if day in schedule and week_type in schedule[day]:
                lessons = schedule[day][week_type]
                if lessons:
                    for lesson_info in lessons:
                        markup = create_attendance_markup(student_id, lesson_info)
                        bot.send_message(chat_id, lesson_info, reply_markup=markup)
                else:
                    bot.send_message(chat_id, f"Сегодня ({day}) нет занятий.")
            else:
                bot.send_message(chat_id, f"Сегодня ({day}) нет занятий.")
        else:
            bot.send_message(chat_id, "Пожалуйста, зарегистрируйтесь сначала с помощью /start.")
    else:
        bot.send_message(chat_id, "Пожалуйста, зарегистрируйтесь сначала с помощью /start.")

# Создание разметки с кнопкой "Был на занятии"
def create_attendance_markup(student_id, lesson_info):
    markup = types.InlineKeyboardMarkup()
    attendance_button = types.InlineKeyboardButton("Был на занятии", callback_data=f"attendance_{student_id}_{lesson_info}")
    markup.add(attendance_button)
    return markup

# Обработчик нажатия кнопки "Был на занятии"
@bot.callback_query_handler(lambda query: query.data.startswith('attendance_'))
def handle_attendance(callback_query):
    callback_data = callback_query.data.split('_')
    student_id = callback_data[1]
    lesson_info = '_'.join(callback_data[2:])
    bot.send_message(callback_query.message.chat.id, f"Студент с номером студенческого билета {student_id} "
                                                     f"отмечен на занятии: {lesson_info}")

def get_current_day():
    return datetime.datetime.now().strftime('%A')

def get_current_week_type():
    current_week_number = datetime.date.today().isocalendar()[1]
    if current_week_number % 2 == 0:
        return 'lower'
    else:
        return 'upper'
    
def load_students_from_sqlite():
    conn = sqlite3.connect('/Users/nkt1lyn/Documents/students.db')
    cursor = conn.cursor()
    cursor.execute('SELECT id, data FROM students')
    result = cursor.fetchall()
    conn.close()
    for row in result:
        student_id = row[0]
        student_data = row[1]
        students[student_id] = {'student_id': student_data}

def save_student_data(student_id, student_data):
    db_path = '/Users/nkt1lyn/Documents/students.db'
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute('INSERT INTO students (id, data) VALUES (?, ?)', (student_id, student_data))
    conn.commit()
    conn.close()

    
import os

def create_database():
    db_path = '/Users/nkt1lyn/Documents/students.db'
    if os.access(db_path, os.R_OK) and os.access(db_path, os.W_OK):
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        cursor.execute('''CREATE TABLE IF NOT EXISTS students
                         (id text PRIMARY KEY, data text)''')
        conn.close()
    else:
        print(f"No read or write permissions for {db_path}")

if __name__ == '__main__':
    create_database()  # Create SQLite database if it doesn't exist
    bot.polling(none_stop=True)
```
